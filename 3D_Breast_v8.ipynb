```python
# First, mount Google Drive to access the files
from google.colab import drive
drive.mount('/content/drive')

import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import cv2
from sklearn.metrics import (confusion_matrix, classification_report, f1_score,
                           precision_score, recall_score, accuracy_score,
                           roc_auc_score, roc_curve, auc, precision_recall_curve,
                           average_precision_score)
import seaborn as sns
import torchvision.transforms as transforms
from torch.optim.lr_scheduler import ReduceLROnPlateau, CosineAnnealingLR
import warnings
warnings.filterwarnings('ignore')

# Advanced imports
from timm.models.layers import DropPath
import torchvision.models as models
from einops import rearrange, repeat

print("‚úÖ All imports completed successfully!")

# ==================== ENHANCED METADATA PROCESSING ====================

def load_and_parse_metadata(metadata_path):
    """Enhanced metadata parsing with better class handling"""
    try:
        # Read the Excel file
        df = pd.read_excel(metadata_path, header=None)
        print(f"üìã Raw metadata loaded with {len(df)} rows")

        # Find where the actual data starts
        data_start_idx = None
        for idx, row in df.iterrows():
            if isinstance(row[0], str) and row[0].startswith('IMG'):
                data_start_idx = idx
                break

        if data_start_idx is None:
            print("‚ùå No image data found in metadata")
            return {}

        # Create proper dataframe
        headers = ['image_ref', 'view', 'tissue', 'abnormality', 'class', 'x', 'y', 'radius']
        df_data = df.iloc[data_start_idx:].copy()
        df_data.columns = headers[:len(df_data.columns)]

        # Enhanced cleaning
        df_clean = df_data.dropna(subset=['image_ref'])
        df_clean = df_clean[df_clean['image_ref'].str.startswith('IMG', na=False)]

        # Parse metadata with enhanced logic
        metadata_dict = {}
        for _, row in df_clean.iterrows():
            img_ref = str(row['image_ref']).strip()

            class_val = str(row['class']).strip().upper() if pd.notna(row['class']) else ""
            abnormality = str(row['abnormality']).upper() if pd.notna(row['abnormality']) else ""
            view = str(row['view']).strip() if pd.notna(row['view']) else ""
            tissue = str(row['tissue']).strip() if pd.notna(row['tissue']) else ""

            # Enhanced classification logic
            if class_val == 'M' or any(malignant_keyword in abnormality for malignant_keyword in ['MALIGNANT', 'SPIC', 'ARCH']):
                final_class = 'M'
            elif class_val == 'B' or 'BENIGN' in abnormality or 'CIRC' in abnormality:
                final_class = 'B'
            elif class_val == 'N' or 'NORM' in abnormality or abnormality == 'NORM':
                final_class = 'N'
            else:
                # Default to normal if no clear indication
                final_class = 'N'

            metadata_dict[img_ref] = {
                'class': final_class,
                'abnormality': abnormality,
                'view': view,
                'tissue': tissue
            }

        print(f"üìä Processed {len(metadata_dict)} metadata entries")

        # Enhanced statistics
        class_counts = {}
        for data in metadata_dict.values():
            cls = data['class']
            class_counts[cls] = class_counts.get(cls, 0) + 1

        print("üìà Enhanced class distribution in metadata:")
        for cls, count in class_counts.items():
            print(f"   {cls}: {count} entries")

        return metadata_dict

    except Exception as e:
        print(f"‚ùå Error loading metadata: {e}")
        import traceback
        traceback.print_exc()
        return {}

# Load metadata
metadata_path = "/content/drive/MyDrive/breast-cancer-mammography/Metadata.xlsx"
metadata = load_and_parse_metadata(metadata_path)

# ==================== ADVANCED TRANSFORMER-CNN HYBRID MODEL ====================

class EnhancedPatchEmbedding(nn.Module):
    """Enhanced patch embedding with positional encoding"""
    def __init__(self, img_size=224, patch_size=16, in_chans=3, embed_dim=768):
        super().__init__()
        self.img_size = img_size
        self.patch_size = patch_size
        self.num_patches = (img_size // patch_size) ** 2

        self.proj = nn.Conv2d(in_chans, embed_dim, kernel_size=patch_size, stride=patch_size)
        self.norm = nn.LayerNorm(embed_dim)

        # Learnable positional encoding
        self.pos_embed = nn.Parameter(torch.zeros(1, self.num_patches + 1, embed_dim))
        self.cls_token = nn.Parameter(torch.zeros(1, 1, embed_dim))

    def forward(self, x):
        B, C, H, W = x.shape
        x = self.proj(x).flatten(2).transpose(1, 2)  # B, N, D
        x = self.norm(x)

        cls_tokens = self.cls_token.expand(B, -1, -1)
        x = torch.cat((cls_tokens, x), dim=1)
        x = x + self.pos_embed

        return x

class MultiScaleTransformerEncoder(nn.Module):
    """Multi-scale transformer encoder for medical images"""
    def __init__(self, dim, depth, heads, mlp_ratio=4., drop_rate=0.):
        super().__init__()
        self.layers = nn.ModuleList()
        for _ in range(depth):
            self.layers.append(nn.ModuleList([
                nn.LayerNorm(dim),
                nn.MultiheadAttention(dim, heads, dropout=drop_rate, batch_first=True),
                nn.LayerNorm(dim),
                nn.Sequential(
                    nn.Linear(dim, int(dim * mlp_ratio)),
                    nn.GELU(),
                    nn.Dropout(drop_rate),
                    nn.Linear(int(dim * mlp_ratio), dim),
                    nn.Dropout(drop_rate)
                )
            ]))

    def forward(self, x):
        for norm1, attn, norm2, ff in self.layers:
            # Self-attention with residual
            x1 = norm1(x)
            attn_out, _ = attn(x1, x1, x1)
            x = x + attn_out

            # Feed-forward with residual
            x2 = norm2(x)
            ff_out = ff(x2)
            x = x + ff_out

        return x

class AdvancedMedicalTransformer(nn.Module):
    """Advanced hybrid model with enhanced transformer integration"""

    def __init__(self, num_classes=3, img_size=224, in_chans=3,
                 embed_dim=512, depth=6, heads=8, mlp_ratio=4.,
                 drop_rate=0.1, cnn_backbone='resnet34'):
        super().__init__()

        # Enhanced CNN Backbone
        if cnn_backbone == 'resnet34':
            backbone = models.resnet34(pretrained=True)
            self.cnn_features = nn.Sequential(
                backbone.conv1,
                backbone.bn1,
                backbone.relu,
                backbone.maxpool,
                backbone.layer1,
                backbone.layer2,
                backbone.layer3,
                backbone.layer4
            )
            cnn_feat_dim = 512
        else:
            # Custom CNN
            self.cnn_features = nn.Sequential(
                nn.Conv2d(in_chans, 64, 7, 2, 3),
                nn.BatchNorm2d(64),
                nn.ReLU(inplace=True),
                nn.MaxPool2d(3, 2, 1),

                nn.Conv2d(64, 128, 3, 1, 1),
                nn.BatchNorm2d(128),
                nn.ReLU(inplace=True),
                nn.MaxPool2d(2, 2),

                nn.Conv2d(128, 256, 3, 1, 1),
                nn.BatchNorm2d(256),
                nn.ReLU(inplace=True),
                nn.MaxPool2d(2, 2),

                nn.Conv2d(256, 512, 3, 1, 1),
                nn.BatchNorm2d(512),
                nn.ReLU(inplace=True),
            )
            cnn_feat_dim = 512

        # Multi-scale feature extraction
        self.feature_pyramid = nn.ModuleDict({
            'high': nn.Conv2d(cnn_feat_dim, embed_dim // 4, 1),
            'medium': nn.Sequential(
                nn.AdaptiveAvgPool2d(28),
                nn.Conv2d(cnn_feat_dim, embed_dim // 4, 1)
            ),
            'low': nn.Sequential(
                nn.AdaptiveAvgPool2d(14),
                nn.Conv2d(cnn_feat_dim, embed_dim // 2, 1)
            )
        })

        # Enhanced Transformer
        self.transformer = MultiScaleTransformerEncoder(
            dim=embed_dim,
            depth=depth,
            heads=heads,
            mlp_ratio=mlp_ratio,
            drop_rate=drop_rate
        )

        # Attention pooling
        self.attention_pool = nn.Sequential(
            nn.Linear(embed_dim, embed_dim // 2),
            nn.Tanh(),
            nn.Linear(embed_dim // 2, 1),
            nn.Softmax(dim=1)
        )

        # Advanced classification head
        self.classifier = nn.Sequential(
            nn.Linear(embed_dim, 512),
            nn.BatchNorm1d(512),
            nn.GELU(),
            nn.Dropout(0.3),

            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.GELU(),
            nn.Dropout(0.2),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.GELU(),
            nn.Dropout(0.1),

            nn.Linear(128, num_classes)
        )

        # Initialize weights
        self.apply(self._init_weights)

    def _init_weights(self, m):
        if isinstance(m, nn.Linear):
            nn.init.trunc_normal_(m.weight, std=.02)
            if m.bias is not None:
                nn.init.constant_(m.bias, 0)
        elif isinstance(m, nn.LayerNorm):
            nn.init.constant_(m.bias, 0)
            nn.init.constant_(m.weight, 1.0)

    def forward(self, x):
        B = x.size(0)

        # CNN feature extraction
        cnn_features = self.cnn_features(x)  # [B, 512, H, W]

        # Multi-scale feature fusion
        high_res = self.feature_pyramid['high'](cnn_features)  # [B, 128, H, W]
        med_res = self.feature_pyramid['medium'](cnn_features)  # [B, 128, 28, 28]
        low_res = self.feature_pyramid['low'](cnn_features)     # [B, 256, 14, 14]

        # Resize and concatenate
        high_res = F.adaptive_avg_pool2d(high_res, 14)
        med_res = F.adaptive_avg_pool2d(med_res, 14)

        fused_features = torch.cat([high_res, med_res, low_res], dim=1)  # [B, 512, 14, 14]
        fused_features = fused_features.flatten(2).transpose(1, 2)  # [B, 196, 512]

        # Transformer processing
        transformer_features = self.transformer(fused_features)  # [B, 196, 512]

        # Attention pooling
        attention_weights = self.attention_pool(transformer_features)  # [B, 196, 1]
        global_features = torch.sum(transformer_features * attention_weights, dim=1)  # [B, 512]

        # Classification
        output = self.classifier(global_features)

        return output

# ==================== ENHANCED DATA PROCESSING ====================

class AdvancedMedicalDataset(Dataset):
    """Enhanced dataset with advanced preprocessing and augmentation"""

    def __init__(self, base_path, target_size=(224, 224), metadata_dict=None, is_training=True):
        self.base_path = Path(base_path)
        self.target_size = target_size
        self.metadata_dict = metadata_dict or {}
        self.is_training = is_training

        # Enhanced transforms
        if is_training:
            self.transform = transforms.Compose([
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.RandomRotation(10),
                transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
                transforms.ColorJitter(brightness=0.2, contrast=0.2),
            ])
        else:
            self.transform = None

        print("üîç Discovering medical images with enhanced metadata labeling...")
        self.samples = self._discover_and_label_images()

        if not self.samples:
            raise ValueError(f"‚ùå No valid medical images found in {base_path}")

        print(f"‚úÖ Final dataset: {len(self.samples)} samples")

        # Enhanced class distribution analysis
        class_dist = pd.Series([s['label_name'] for s in self.samples]).value_counts()
        print("üìä Enhanced class distribution:")
        for cls, count in class_dist.items():
            print(f"   {cls}: {count} samples")

    def _discover_and_label_images(self):
        """Enhanced image discovery with better metadata matching"""
        image_extensions = {'.tif', '.tiff', '.png', '.jpg', '.jpeg', '.bmp'}
        all_images = []

        # Enhanced recursive search
        for ext in image_extensions:
            images = list(self.base_path.rglob(f'*{ext}'))
            images.extend(list(self.base_path.rglob(f'*{ext.upper()}')))
            all_images.extend(images)

        print(f"üì∏ Found {len(all_images)} potential medical images")

        # Enhanced filtering
        valid_images = []
        seen_files = set()

        for img_path in all_images:
            if img_path.name.lower() in seen_files:
                continue
            seen_files.add(img_path.name.lower())

            filename = img_path.stem.lower()
            # Enhanced filtering criteria
            if any(keyword in filename for keyword in ['mask', 'roi', 'segmentation', 'label', 'thumbnail', 'thumb', 'copy']):
                continue

            try:
                with Image.open(img_path) as img:
                    img.verify()
                valid_images.append(img_path)
            except Exception as e:
                continue

        print(f"üì∑ Validated {len(valid_images)} medical images")

        # Enhanced metadata matching
        samples = []
        for img_path in valid_images:
            label, label_name = self._get_label_from_metadata(img_path)
            if label is not None:
                samples.append({
                    'image_path': str(img_path),
                    'label': label,
                    'label_name': label_name,
                    'image_id': img_path.stem,
                    'file_path': str(img_path)
                })

        return self._enhanced_balance_dataset(samples)

    def _get_label_from_metadata(self, image_path):
        """Enhanced metadata matching with fallback strategies"""
        image_id = Path(image_path).stem.upper()

        # Multiple matching strategies
        matching_strategies = [
            # Exact match
            lambda: self.metadata_dict.get(image_id),
            # Remove suffixes
            lambda: self.metadata_dict.get(image_id.split('_')[0]),
            # Remove -COPY etc
            lambda: self.metadata_dict.get(image_id.replace('-COPY', '').replace('_COPY', '')),
            # Partial matching
            lambda: next((v for k, v in self.metadata_dict.items() if image_id in k or k in image_id), None)
        ]

        for strategy in matching_strategies:
            metadata = strategy()
            if metadata:
                class_val = metadata.get('class', '')
                abnormality = metadata.get('abnormality', '')

                if class_val == 'M' or any(malignant_keyword in abnormality for malignant_keyword in ['MALIGNANT', 'SPIC', 'ARCH']):
                    return 0, 'malignant'
                elif class_val == 'B' or 'BENIGN' in abnormality or 'CIRC' in abnormality:
                    return 1, 'benign'
                elif class_val == 'N' or 'NORM' in abnormality or not abnormality:
                    return 2, 'normal'

        return None, 'unknown'

    def _enhanced_balance_dataset(self, samples):
        """Enhanced balancing with class-aware sampling"""
        if len(samples) < 10:
            return samples

        class_groups = {0: [], 1: [], 2: []}

        for sample in samples:
            if sample['label'] in class_groups:
                class_groups[sample['label']].append(sample)

        # Enhanced balancing strategy
        class_sizes = {k: len(v) for k, v in class_groups.items() if len(v) > 0}
        if not class_sizes:
            return samples

        # Use median size for better balance
        sizes = list(class_sizes.values())
        target_size = int(np.median(sizes))

        balanced_samples = []
        for label, group in class_groups.items():
            if len(group) > 0:
                if len(group) < target_size:
                    # Oversample minority classes
                    repeat_times = target_size // len(group) + 1
                    augmented_group = group * repeat_times
                    balanced_samples.extend(augmented_group[:target_size])
                else:
                    # Undersample majority classes
                    np.random.shuffle(group)
                    balanced_samples.extend(group[:target_size])

        print(f"‚öñÔ∏è Enhanced balancing: {len(samples)} ‚Üí {len(balanced_samples)} samples")
        return balanced_samples

    def _enhanced_medical_image_loading(self, image_path):
        """Enhanced medical image loading with advanced preprocessing"""
        try:
            # Load with PIL
            with Image.open(image_path) as img:
                if img.mode != 'RGB':
                    img = img.convert('RGB')
                img_array = np.array(img, dtype=np.float32)

            # Enhanced normalization
            if img_array.max() > 255:
                img_array = img_array / 65535.0
            else:
                img_array = img_array / 255.0

            # Enhanced preprocessing
            img_array = self._advanced_preprocessing(img_array)

            # Resize with enhanced interpolation
            if img_array.shape[:2] != self.target_size:
                img_pil = Image.fromarray((img_array * 255).astype(np.uint8))
                img_pil = img_pil.resize((self.target_size[1], self.target_size[0]),
                                       Image.LANCZOS)
                img_array = np.array(img_pil, dtype=np.float32) / 255.0

            # Apply training transforms if available
            if self.transform and self.is_training:
                img_pil = Image.fromarray((img_array * 255).astype(np.uint8))
                img_pil = self.transform(img_pil)
                img_array = np.array(img_pil, dtype=np.float32) / 255.0

            # Enhanced normalization
            img_array = (img_array - np.mean(img_array)) / (np.std(img_array) + 1e-8)
            volume = np.transpose(img_array, (2, 0, 1))  # [C, H, W]

            return volume

        except Exception as e:
            print(f"‚ùå Error loading {image_path}: {e}")
            # Enhanced fallback
            return np.random.rand(3, self.target_size[0], self.target_size[1]).astype(np.float32) * 0.1

    def _advanced_preprocessing(self, image):
        """Advanced medical image preprocessing"""
        # Contrast enhancement
        p2, p98 = np.percentile(image, (2, 98))
        if p98 > p2:
            image = np.clip((image - p2) / (p98 - p2), 0, 1)

        # Noise reduction (simple median filter)
        if len(image.shape) == 3:
            for i in range(image.shape[2]):
                image[:,:,i] = cv2.medianBlur((image[:,:,i] * 255).astype(np.uint8), 3) / 255.0
        else:
            image = cv2.medianBlur((image * 255).astype(np.uint8), 3) / 255.0

        return image

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        sample = self.samples[idx]

        volume = self._enhanced_medical_image_loading(sample['image_path'])
        volume_tensor = torch.from_numpy(volume).float()

        return {
            'volume': volume_tensor,
            'label': torch.tensor(sample['label'], dtype=torch.long),
            'label_name': sample['label_name'],
            'image_id': sample['image_id'],
            'image_path': sample['image_path']
        }

# ==================== ADVANCED TRAINING STRATEGY ====================

class AdvancedMedicalTrainer:
    def __init__(self, model, config, device):
        self.model = model
        self.config = config
        self.device = device

        # Enhanced loss function with class weights
        class_weights = torch.tensor([1.0, 1.0, 1.0]).to(device)  # Can be adjusted based on class distribution
        self.criterion = nn.CrossEntropyLoss(weight=class_weights)

        # Enhanced optimizer with different learning rates for backbone and head
        backbone_params = []
        head_params = []
        for name, param in model.named_parameters():
            if 'cnn_features' in name:
                backbone_params.append(param)
            else:
                head_params.append(param)

        self.optimizer = torch.optim.AdamW([
            {'params': backbone_params, 'lr': config.learning_rate * 0.1},
            {'params': head_params, 'lr': config.learning_rate}
        ], weight_decay=1e-4)

        # Enhanced scheduler
        self.scheduler = CosineAnnealingLR(self.optimizer, T_max=config.num_epochs)

        # Gradient clipping
        self.grad_clip = 1.0

        self.best_accuracy = 0
        self.history = {
            'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': [],
            'learning_rates': []
        }

    def train_epoch(self, train_loader):
        self.model.train()
        total_loss = 0
        correct = 0
        total = 0

        pbar = tqdm(train_loader, desc="üöÄ Advanced Training")
        for batch in pbar:
            volumes = batch['volume'].to(self.device)
            labels = batch['label'].to(self.device)

            self.optimizer.zero_grad()
            outputs = self.model(volumes)
            loss = self.criterion(outputs, labels)
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), self.grad_clip)

            self.optimizer.step()

            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            pbar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Acc': f'{100 * correct / total:.2f}%'
            })

        epoch_loss = total_loss / len(train_loader)
        epoch_acc = 100 * correct / total

        return epoch_loss, epoch_acc

    def validate_epoch(self, val_loader):
        self.model.eval()
        total_loss = 0
        correct = 0
        total = 0

        with torch.no_grad():
            for batch in tqdm(val_loader, desc="üß™ Advanced Validating"):
                volumes = batch['volume'].to(self.device)
                labels = batch['label'].to(self.device)

                outputs = self.model(volumes)
                loss = self.criterion(outputs, labels)

                total_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        epoch_loss = total_loss / len(val_loader)
        epoch_acc = 100 * correct / total

        return epoch_loss, epoch_acc

    def train(self, train_loader, val_loader, save_path):
        print("üöÄ Starting Advanced Medical AI Training...")

        for epoch in range(self.config.num_epochs):
            print(f"\nüìç Epoch {epoch+1}/{self.config.num_epochs}")
            print("-" * 50)

            # Training
            train_loss, train_acc = self.train_epoch(train_loader)

            # Validation
            val_loss, val_acc = self.validate_epoch(val_loader)

            # Update scheduler
            self.scheduler.step()
            current_lr = self.optimizer.param_groups[0]['lr']

            # Update history
            self.history['train_loss'].append(train_loss)
            self.history['val_loss'].append(val_loss)
            self.history['train_acc'].append(train_acc)
            self.history['val_acc'].append(val_acc)
            self.history['learning_rates'].append(current_lr)

            print(f"üìä Train - Loss: {train_loss:.4f}, Acc: {train_acc:.2f}%")
            print(f"üìä Val   - Loss: {val_loss:.4f}, Acc: {val_acc:.2f}%")
            print(f"üìà Learning Rate: {current_lr:.2e}")

            # Enhanced model saving
            if val_acc > self.best_accuracy:
                self.best_accuracy = val_acc
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': self.model.state_dict(),
                    'optimizer_state_dict': self.optimizer.state_dict(),
                    'best_accuracy': self.best_accuracy,
                    'history': self.history
                }, save_path)
                print(f"üíæ Saved best model with val_acc: {val_acc:.2f}%")

        return self.history

# ==================== ENHANCED XAI VISUALIZATION ====================

class AdvancedXAIVisualizer:
    def __init__(self, model, device):
        self.model = model
        self.device = device
        self.integrated_gradients = IntegratedGradients(model)
        self.saliency = Saliency(model)

    def comprehensive_visualization(self, sample, class_names=['Malignant', 'Benign', 'Normal']):
        self.model.eval()

        input_tensor = sample['volume'].unsqueeze(0).to(self.device)
        true_label = sample['label'].item()

        with torch.no_grad():
            output = self.model(input_tensor)
            pred_probs = F.softmax(output, dim=1)
            pred_label = torch.argmax(output, dim=1).item()
            confidence = pred_probs[0, pred_label].item()

        print(f"üìä Sample: {sample['image_id']}")
        print(f"   True: {class_names[true_label]}, Predicted: {class_names[pred_label]} (Confidence: {confidence:.2%})")
        print(f"   Class Probabilities: Malignant: {pred_probs[0,0]:.2%}, Benign: {pred_probs[0,1]:.2%}, Normal: {pred_probs[0,2]:.2%}")

        try:
            # Multiple attribution methods
            # 1. Saliency Maps
            saliency_attr = self.saliency.attribute(input_tensor, target=pred_label)
            saliency_attr = saliency_attr.squeeze().cpu().detach().numpy()
            saliency_attr = np.mean(saliency_attr, axis=0)

            # 2. Integrated Gradients
            baseline = torch.zeros_like(input_tensor)
            ig_attr = self.integrated_gradients.attribute(input_tensor, baseline, target=pred_label)
            ig_attr = ig_attr.squeeze().cpu().detach().numpy()
            ig_attr = np.mean(ig_attr, axis=0)

            # Original image
            original_image = input_tensor.squeeze().cpu().detach().numpy()
            original_image = np.mean(original_image, axis=0)

            # Enhanced visualization
            fig, axes = plt.subplots(2, 2, figsize=(12, 10))

            # Original image
            axes[0,0].imshow(original_image, cmap='gray')
            axes[0,0].set_title(f'Original Image\nTrue: {class_names[true_label]}', fontweight='bold')
            axes[0,0].axis('off')

            # Saliency map
            im1 = axes[0,1].imshow(saliency_attr, cmap='hot')
            axes[0,1].set_title(f'Saliency Map\nPred: {class_names[pred_label]}', fontweight='bold')
            axes[0,1].axis('off')
            plt.colorbar(im1, ax=axes[0,1], fraction=0.046, pad=0.04)

            # Integrated Gradients
            im2 = axes[1,0].imshow(ig_attr, cmap='coolwarm')
            axes[1,0].set_title('Integrated Gradients', fontweight='bold')
            axes[1,0].axis('off')
            plt.colorbar(im2, ax=axes[1,0], fraction=0.046, pad=0.04)

            # Overlay
            axes[1,1].imshow(original_image, cmap='gray')
            im3 = axes[1,1].imshow(ig_attr, cmap='hot', alpha=0.5)
            axes[1,1].set_title('Attention Overlay', fontweight='bold')
            axes[1,1].axis('off')
            plt.colorbar(im3, ax=axes[1,1], fraction=0.046, pad=0.04)

            plt.tight_layout()
            plt.show()

            return {
                'true_label': true_label,
                'pred_label': pred_label,
                'confidence': confidence,
                'probabilities': pred_probs[0].cpu().numpy()
            }

        except Exception as e:
            print(f"‚ö†Ô∏è Advanced XAI failed: {e}")
            # Fallback to simple visualization
            original_image = input_tensor.squeeze().cpu().detach().numpy()
            original_image = np.mean(original_image, axis=0)

            plt.figure(figsize=(6, 4))
            plt.imshow(original_image, cmap='gray')
            plt.title(f'Original Image\nTrue: {class_names[true_label]}, Pred: {class_names[pred_label]} ({confidence:.2%})', fontweight='bold')
            plt.axis('off')
            plt.tight_layout()
            plt.show()

            return {
                'true_label': true_label,
                'pred_label': pred_label,
                'confidence': confidence,
                'probabilities': pred_probs[0].cpu().numpy()
            }

# ==================== ADVANCED PIPELINE ====================

def execute_advanced_medical_pipeline():
    """Advanced medical AI pipeline with enhanced transformer integration"""
    print("üè• ADVANCED MEDICAL AI PIPELINE")
    print("üî¨ Enhanced Transformer ‚Ä¢ Multi-scale Features ‚Ä¢ Advanced Training")
    print("=" * 80)

    class AdvancedConfig:
        def __init__(self):
            self.base_path = "/content/drive/MyDrive/breast-cancer-mammography"
            self.save_dir = "./advanced_medical_ai"
            os.makedirs(self.save_dir, exist_ok=True)

            # Enhanced model parameters
            self.image_size = 224
            self.num_classes = 3
            self.embed_dim = 512
            self.transformer_depth = 6
            self.transformer_heads = 8
            self.mlp_ratio = 4.0
            self.drop_rate = 0.1

            # Advanced training parameters
            self.batch_size = 8
            self.num_epochs = 25
            self.learning_rate = 2e-4

            self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    config = AdvancedConfig()
    print(f"üíª Using device: {config.device}")
    print(f"üìÅ Data path: {config.base_path}")
    print(f"üìã Using enhanced metadata: {len(metadata)} entries")

    try:
        # Enhanced dataset loading
        print("\nüìä STEP 1: LOADING ENHANCED DATASET")
        train_dataset = AdvancedMedicalDataset(config.base_path, (config.image_size, config.image_size), metadata, is_training=True)
        val_dataset = AdvancedMedicalDataset(config.base_path, (config.image_size, config.image_size), metadata, is_training=False)
        test_dataset = AdvancedMedicalDataset(config.base_path, (config.image_size, config.image_size), metadata, is_training=False)

        # Create data loaders
        train_loader = DataLoader(train_dataset, batch_size=min(config.batch_size, len(train_dataset)), shuffle=True, num_workers=2)
        val_loader = DataLoader(val_dataset, batch_size=min(config.batch_size, len(val_dataset)), shuffle=False, num_workers=2)
        test_loader = DataLoader(test_dataset, batch_size=min(config.batch_size, len(test_dataset)), shuffle=False, num_workers=2)

        print(f"‚úÖ Training: {len(train_dataset)} samples")
        print(f"‚úÖ Validation: {len(val_dataset)} samples")
        print(f"‚úÖ Test: {len(test_dataset)} samples")

        # Create advanced model
        print("\nüèóÔ∏è STEP 2: CREATING ADVANCED TRANSFORMER MODEL")
        model = AdvancedMedicalTransformer(
            num_classes=config.num_classes,
            img_size=config.image_size,
            embed_dim=config.embed_dim,
            depth=config.transformer_depth,
            heads=config.transformer_heads,
            mlp_ratio=config.mlp_ratio,
            drop_rate=config.drop_rate,
            cnn_backbone='resnet34'
        )
        model = model.to(config.device)

        total_params = sum(p.numel() for p in model.parameters())
        trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
        print(f"üìà Advanced Model Parameters: {total_params:,} total, {trainable_params:,} trainable")

        # Advanced training
        print("\nüöÄ STEP 3: ADVANCED TRAINING")
        save_path = os.path.join(config.save_dir, "advanced_best_model.pth")
        trainer = AdvancedMedicalTrainer(model, config, config.device)
        history = trainer.train(train_loader, val_loader, save_path)

        # Enhanced evaluation
        print("\nüìä STEP 4: ENHANCED EVALUATION")
        checkpoint = torch.load(save_path)
        model.load_state_dict(checkpoint['model_state_dict'])
        model.eval()

        test_preds, test_labels, test_probs = [], [], []
        with torch.no_grad():
            for batch in tqdm(test_loader, desc="üß™ Advanced Testing"):
                volumes = batch['volume'].to(config.device)
                labels = batch['label'].to(config.device)

                outputs = model(volumes)
                probs = F.softmax(outputs, dim=1)
                preds = torch.argmax(outputs, dim=1)

                test_preds.extend(preds.cpu().numpy())
                test_labels.extend(labels.cpu().numpy())
                test_probs.extend(probs.cpu().numpy())

        test_probs = np.array(test_probs)

        # Generate comprehensive report
        metrics_calculator = ComprehensiveMedicalMetrics()
        final_metrics = metrics_calculator.generate_report(test_labels, test_preds, test_probs, history)

        # Advanced XAI
        print("\nüîç STEP 5: ADVANCED XAI VISUALIZATION")
        xai_visualizer = AdvancedXAIVisualizer(model, config.device)

        sample_indices = min(3, len(test_dataset))
        print(f"\nüìã Advanced XAI Analysis on {sample_indices} Sample Predictions:")

        for i in range(sample_indices):
            sample_idx = torch.randint(0, len(test_dataset), (1,)).item()
            sample = test_dataset[sample_idx]
            xai_results = xai_visualizer.comprehensive_visualization(sample)

        # Final enhanced summary
        print("\n" + "=" * 80)
        print("üéâ ADVANCED MEDICAL AI PIPELINE COMPLETED SUCCESSFULLY!")
        print("=" * 80)
        print(f"üìä Final Test Accuracy: {final_metrics['accuracy']:.4f}")
        print(f"üìä Final Test F1-Score: {final_metrics['f1_score']:.4f}")
        print(f"üìä Final Test AUC: {final_metrics['auc']:.4f}")
        print(f"üèÜ Best Validation Accuracy: {checkpoint['best_accuracy']:.2f}%")
        print(f"üíæ Advanced Model Saved: {save_path}")

        return {
            'model': model,
            'metrics': final_metrics,
            'history': history,
            'checkpoint': checkpoint
        }

    except Exception as e:
        print(f"‚ùå Error in advanced pipeline: {e}")
        import traceback
        traceback.print_exc()
        return None

# ==================== EXECUTE ADVANCED PIPELINE ====================

if __name__ == "__main__":
    print("üöÄ LAUNCHING ADVANCED MEDICAL AI PIPELINE")
    print("üìã Enhanced Transformer ‚Ä¢ Multi-scale Features ‚Ä¢ Advanced Training")

    results = execute_advanced_medical_pipeline()

    if results is not None:
        print("‚úÖ Advanced pipeline completed successfully!")
        print(f"üèÜ Best Model Accuracy: {results['metrics']['accuracy']:.4f}")
        print(f"üéØ Key Improvements:")
        print(f"   ‚Ä¢ Enhanced transformer architecture")
        print(f"   ‚Ä¢ Multi-scale feature fusion")
        print(f"   ‚Ä¢ Advanced training strategies")
        print(f"   ‚Ä¢ Comprehensive XAI visualization")
    else:
        print("‚ùå Advanced pipeline failed. Please check the error messages above.")
```
